```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stocks = ['AAPL','AMZN','BA','CAT','CVX','GOOGL','GS','JNJ','JPM','KO','MCD','MSFT','NKE','NVDA','PFE','SOFI','TSLA','UNH','WMT','XOM']
data = pd.read_csv("Trading_Project_Data.csv", header=[0,1], index_col =0)
data.drop('Dividends', axis=1, inplace=True)
data.drop('Stock Splits', axis=1, inplace=True)
data = data.swaplevel(0,1,axis=1).sort_index(axis=1)
data.dropna(inplace=True)
tickers =data.columns.get_level_values(0).unique().tolist()
data = data.loc['1/3/2024':]
data
```

```python
# obtaining more training data using yfinance 
new = yf.Tickers(stocks)
new_data = new.history(start = '2023-01-17', end = '2024-03-01')
new_data.drop('Dividends', axis=1, inplace=True)
new_data.drop('Stock Splits', axis=1, inplace=True)
new_data = new_data.swaplevel(0,1,axis=1).sort_index(axis=1)
new_data
```

# Identifying Support and Resistance Levels

```python

def identify_levels(data, window):
    data = data.copy()
    data["Support"] = data["Low"].rolling(window=window).min().shift()
    data["Resistance"] = data["High"].rolling(window=window).max().shift()
    data.dropna(inplace = True)
    return data
```

# Detect Breakouts

```python

def detect_breakouts(data):
    data = data.copy()
    data["Bullish_Breakout"] = (data["Close"].copy().astype(float) > data["Resistance"])
    data["Bearish_Breakout"] = (data["Close"].copy().astype(float) < data["Support"])
    data.dropna(inplace = True)
    return data
```

# Defining Position

```python
def pos(data):
    data = data.copy()

    # Initialize 'pos' column with 0
    data["pos"] = 0.0  

    # Vectorized conditional assignments
    data.loc[data["Bullish_Breakout"] & ~data["Bearish_Breakout"], "pos"] = 1.0
    data.loc[~data["Bullish_Breakout"] & data["Bearish_Breakout"], "pos"] = -0.5

    return data
```

# Computing Strategy Returns

```python
def strat_returns(stock, window, df): 
    stock_data = df.xs(stock, axis=1, level=0)

    stock_data = identify_levels(stock_data, window) 
    stock_data = detect_breakouts(stock_data)
    stock_data = pos(stock_data)
    stock_data.dropna(inplace = True) 
    
    stock_data['Returns'] = np.log(stock_data['Close'].astype(float)/stock_data['Close'].astype(float).shift())
    stock_data.fillna(0, inplace = True)
    stock_data['Strat_Returns'] = stock_data['Returns']*stock_data['pos'].shift()
    #print('Total Returns =' ,np.exp(data['Strat_Returns'].sum()))
    ret = stock_data[['Returns','Strat_Returns']]
    return ret
```

# Varying Rolling Days

```python
res = pd.DataFrame()
for stock in stocks:
    res[(f'{stock}','strat_ret')] =0.0
    res[(f'{stock}','Benchmark')] =0.0
    res[(f'{stock}','Outperform')] =0.0
    res.columns = pd.MultiIndex.from_tuples(res.columns)
    for i in [5,10,15,20, 25,28]: 
        ret = strat_returns(stock, i, new_data) 
        total_ret = float(np.exp(ret['Strat_Returns'].sum()))
        benchmark = float(np.exp(ret['Returns'].sum()))
        res.loc[f'{i}',(f'{stock}','strat_ret')] = total_ret
        res.loc[f'{i}',(f'{stock}','Benchmark')] = benchmark
        res.loc[f'{i}',(f'{stock}','Outperform')] = total_ret - benchmark 

res
```

# Finding Best Outperformance per Stock

```python
m = res.max(axis =0) # returns axis with best strat_returns 
m.xs('Outperform',level =1).sort_values(ascending=False) # seeing outperformance levels only 
```
