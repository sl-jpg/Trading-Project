```python
# sharpe

def get_Sharpe(data,rf):
  sharpe= pd.DataFrame(index = stocks, columns = ['Sharpe']) 
  for stock in stocks:
      close = data[[(stock,'Close')]]
      log_ret = np.log(close[[(stock,'Close')]]/close[[(stock,'Close')]].shift())
      mu = log_ret.mean()*252
      sigma = log_ret.std()*np.sqrt(252)
      ssharpe = (mu-rf)/sigma
      sharpe.loc[stock,'Sharpe'] = ssharpe.values
  return sharpe.sort_values(by = 'Sharpe', ascending= False)
```

```python
def rolling_sharpe(data,rf, window): 
    sharpe= pd.DataFrame(index = stocks, columns = ['Sharpe']) 
    for stock in stocks:
      close = data[[(stock,'Close')]]
      log_ret = np.log(close[[(stock,'Close')]]/close[[(stock,'Close')]].shift())
      mu = log_ret.rolling(window = window).mean()*252
      sigma = log_ret.rolling(window = window).std()*np.sqrt(252)
      ssharpe = (mu-np.log(1+rf))/sigma
      sharpe.loc[stock,'Sharpe'] = ssharpe.mean().values
    return sharpe.sort_values(by = 'Sharpe', ascending= False)
```

```python
def rolling_vol(data, window): 
    volume= pd.DataFrame(index = stocks, columns = ['Volume'])
    for stock in stocks:
        vol = data[[(stock,'Volume')]]
        rolling_mean_vol = vol.rolling(window = window).mean()
        volume.loc[stock,'Volume'] = rolling_mean_vol.mean().values
    return volume.sort_values(by='Volume', ascending = False)
```

```python

# Identify Support and Resistance Levels
def identify_levels(data, window):
    data = data.copy()
    data["Support"] = data["Low"].rolling(window=window).min().shift()
    data["Resistance"] = data["High"].rolling(window=window).max().shift()
    data.dropna(inplace = True)
    return data
```

```python
# Detect Breakouts
def detect_breakouts(data):
    data = data.copy()
    data["Bullish_Breakout"] = (data["Close"].copy().astype(float) > data["Resistance"])
    data["Bearish_Breakout"] = (data["Close"].copy().astype(float) < data["Support"])
    data.dropna(inplace = True)
    return data
```

```python
def pos(data):
    data = data.copy()

    # Initialize 'pos' column with 0
    data["pos"] = 0.0  

    # Vectorized conditional assignments
    data.loc[data["Bullish_Breakout"] & ~data["Bearish_Breakout"], "pos"] = 1.0
    data.loc[~data["Bullish_Breakout"] & data["Bearish_Breakout"], "pos"] = -0.5

    return data
```

```python
def strat_returns(stock, window, df): 
    stock_data = df.xs(stock, axis=1, level=0)

    stock_data = identify_levels(stock_data, window) 
    stock_data = detect_breakouts(stock_data)
    stock_data = pos(stock_data)
    stock_data.dropna(inplace = True) 
    
    stock_data['Returns'] = np.log(stock_data['Close'].astype(float)/stock_data['Close'].astype(float).shift())
    stock_data.fillna(0, inplace = True)
    stock_data['Strat_Returns'] = stock_data['Returns']*stock_data['pos'].shift()
    #print('Total Returns =' ,np.exp(data['Strat_Returns'].sum()))
    ret = stock_data[['Returns','Strat_Returns']]
    return ret
```

```python
' using bollinger bands and ATR '

st_dev = 1.5 # 95% CI 
def calculate_indicators(data, boll_window, atr_window):
    data['SMA'] = data['Close'].rolling(boll_window).mean()
    data['Upper_Band'] = data['SMA'] + (st_dev * data['Close'].rolling(boll_window).std())
    data['Lower_Band'] = data['SMA'] - (st_dev * data['Close'].rolling(boll_window).std())
    data['ATR'] = data['High'].rolling(atr_window).max() - data['Low'].rolling(atr_window).min()
    return data
calculate_indicators(get_stock('GS'), 5, 5)
```

```python
def breakout_strategy(stock_data,boll_window,atr_window):
    stock_data = calculate_indicators(stock_data, boll_window, atr_window)
    stock_data['Pos'] = 0  # No trade by default
    stock_data['Stop_Loss'] = np.zeros((len(stock_data), 1))
    stock_data['Take_Profit'] = np.zeros((len(stock_data), 1))
    stock_data['log_ret'] = np.log(stock_data['Close'].astype(float)/ stock_data['Close'].astype(float).shift())

    for i in range(1, len(stock_data)):
        atr = stock_data['ATR'][i-1]
        close = float(stock_data['Close'].iloc[i-1])
        if close > stock_data['Upper_Band'].iloc[i-1]:
            stock_data['Pos'].iloc[i] = 1  # Long
            stock_data['Stop_Loss'].iloc[i] = close - (1.5 * atr)
            stock_data['Take_Profit'].iloc[i] = close + (2 * atr)
            
        elif close < stock_data['Lower_Band'].iloc[i]:
            stock_data['Pos'].iloc[i] = -1  # Short
            stock_data['Stop_Loss'].iloc[i] = close + (1.5 * atr)
            stock_data['Take_Profit'].iloc[i] = close - (2 * atr)
    stock_data['strat_ret'] = stock_data['log_ret']*stock_data['Pos']
    return stock_data
```

```python
def boll_ret(strategy_df):
    print(np.exp(strategy_df['strat_ret'].sum()))
```

```python
boll_ret(breakout_strategy(get_stock('AAPL'),5,5))
```
